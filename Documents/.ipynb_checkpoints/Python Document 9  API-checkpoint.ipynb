```python
# we only want the Driver Standings
my_racer_data = response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']
my_racer_data
```

```python
# what if we only want the last names of the first 5 drivers
for x in range(5):
    data = response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings'][x]['Driver']['familyName']
    print(data)
```

```python
def get_racer_info(data):
    new_data = []
    for racer in data:
        racer_dict = {}
        racer_name = f"{racer['Driver']['givenName']} {racer['Driver']['familyName']}"
        
        racer_dict[racer_name] = {
            'first_name': racer['Driver']['givenName'],
            'last_name': racer['Driver']['familyName'],
            'position': racer['position'],
            'wins': racer['wins'],
            'DOB': racer['Driver']['dateOfBirth'],
            'nationality': racer['Driver']['nationality'],
            'constructor': racer['Constructors'][0]['name']
        }
        new_data.append(racer_dict)
    return new_data
    
get_racer_info(my_racer_data)
```

```python
def get_driver_info_by_year_and_round(year, r):
    url = f"https://ergast.com/api/f1/{year}/{r}/driverStandings.json"
    response = requests.get(url)
    if not response.json()["MRData"]["StandingsTable"]["StandingsLists"]:
        return 'We had an error loading your data. It is likely the year or round is not in the database'
    my_driver_info = []
    for x in range(5):
        data = response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings'][x]
        my_driver_info.append(data)
    return my_driver_info

get_driver_info_by_year_and_round(2019, 4)
```

```python
""" HomeWork
check out this Pokemon API https://pokeapi.co/ Use the requested package to connect this API and get and store data
for 5 different pokemon. Get the pokemons: name, atleast one ability's name, base_experience, and the URL for its 
sprite (an image that shows up on screen) for the 'front_shiny', attack base_state, hp base_stat, defense base_stat"""
```

```python
import requests, json
pokemon_list = ['piplup', 'turtwig', 'chimchar', 'metagross','mew']
pokemon_saved_info = []
def pokemon_info(pokemon):
    url = f'https://pokeapi.co/api/v2/pokemon/{pokemon}'
    pokemon_API = requests.get(url)
    pokemon_data = pokemon_API.json()
    pokemon_dict[pokemon] = {
        'name' : pokemon_data['forms'][0]['name'],
        'first_ability': pokemon_data['abilities'][0]['ability']['name'],
        'base_experience': pokemon_data['base_experience'],
        'front_shiny_sprite_url': pokemon_data['sprites']['front_shiny'],
        'base_attack_stat': pokemon_data['stats'][1]['base_stat'],
        'base_hp_stat': pokemon_data['stats'][0]['base_stat'],
        'base_defence_stat': pokemon_data['stats'][2]['base_stat'],
    }
    pokemon_saved_info.append(pokemon_dict[pokemon])
    
for pokemon in pokemon_list:
    pokemon_info(pokemon)

pokemon_saved_info
```
